```python
from selenium import webdriver as wb #webdriver는 웹을 통제하기 위한 모듈
from selenium.webdriver.common.keys import Keys #Keys는 웹을 제어하면서 값을 입력하기 위한 모듈
import time
import lxml #파싱
import requests as req
from bs4 import BeautifulSoup as bs
import os #파일, 폴더를 생성, 삭제, 존재여부를 알수 있는 라이브러리
from urllib.request import urlretrieve #이미지 경로를 파일로 저장
import pyautogui

driver =wb.Chrome() #실행시키면 크롬 빈창이 나옴, 크롬드라이버 위치 확인 (현재 실행파일이 드라이버와 같은 위치에 있다 가정하에 코드를 구현)
url = "https://charat.me/chibi/create/" #url 변수에 가져오고 싶은 페이지 주소 문자형태로 넣은 것 뿐 
driver.get(url)  #driver변수에 url집어 넣어줌

#오류나도 계속 실행하게끔
while True:
    try:
        button1 = driver.find_element_by_id('emenubtnMenuclick')
        button1.click()
        time.sleep(1)
        button2 = driver.find_element_by_id('elistbtnRandom')
        button2.click()
        time.sleep(1)
        button5 = driver.find_element_by_id('chibiCategoryBtn7')
        button5.click()
        time.sleep(1)
        pyautogui.moveTo(637, 486) #마우스를 x,y좌표로 이동시켜줌
        pyautogui.scroll(-10) #마우스 스크롤 내려줌
        time.sleep(1)
        button6 = driver.find_element_by_id('ccolor_e40613')
        button6.click()
        time.sleep(1)
        button7 = driver.find_element_by_id('chibiCategoryBtn9')
        button7.click()
        time.sleep(1)
        button6.click()
        time.sleep(1)
        button8 = driver.find_element_by_id('chibiCategoryBtn10')
        button8.click()
        time.sleep(1)
        button6.click()
        time.sleep(1)
        pyautogui.moveTo(1105, 787) #마우스를 지정된 x,y좌표로 이동시켜줌
        pyautogui.scroll(-10) #마우스 스크롤 내려줌
        time.sleep(1)
        button9 = driver.find_element_by_id('chibiCategoryBtn11')
        button9.click()
        time.sleep(1)
        button6.click()
        time.sleep(1)
        button3 = driver.find_element_by_id('efloatbtnSaveChara')
        button3.click()
        time.sleep(1)
        button4 = driver.find_element_by_id('chaMakeResultBtnDL')
        button4.click()
        time.sleep(1)
        driver.refresh()
        time.sleep(4)
    except: #위에 try문에서 오류가 난다면
        driver.refresh() #홈페이지 새로고침
        time.sleep(8) #코드를 8초 쉬었다가 다시 처음부터 코드가 while문에 의해 첫번째인 try문으로 들어감
```
